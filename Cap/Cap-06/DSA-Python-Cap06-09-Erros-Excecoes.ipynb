### Try, Except, Finally

```python
try:
    8 + 's'
except TypeError:
    print('Amigão como você soma 8 com S pae?')
```

```python
try:
    f = open('C:\\PythonDSA\\Learn-Python\\Cap\\Cap-06\\arquivos\\arquivo2.txt', 'w')
    f.write('Gravando no arquivo')
except IOError:
    print('Erro: arquivo não encontrado ou não pode ser salvo')
else:
    print("Conteúdo gravado com sucesso!")
```

```python
try:
    f = open('C:\\PythonDSA\\Learn-Python\\Cap\\Cap-06\arquivos\\arquivo2.', 'r')
except IOError:
    print('Erro: arquivo não encontrado ou não pode ser salvo')
else:
    print("Conteúdo gravado com sucesso!")
    f.close()
```

```python
# Utilizando try, except, else, finally
try:
    f = open('C:\\PythonDSA\\Learn-Python\\Cap\\Cap-06\\arquivos\\arquivo2.txt', 'w')
    f.write('Gravando no arquivo')
except IOError:
    print('Erro: arquivo não encontrado ou não pode ser salvo.')
else:
    print('Conteúdo gravado com sucesso!')
    f.close()
finally:
    print('Comandos no blocos finally são sempre executados!')
```

##### Cada possibilidade de erro deve ser tratada no seu programa

```python
def askint():
    try:
        val = int(input("Digite um número: "))
    except:
        print("Você não digitou um número!")
    finally:
        print("Obrigado!")
```

```python
askint()
```

```python
def askit():
    try:
        val = int(input("Digite um número: "))
    except:
        print("Você não digitou um número!")
        val = int(input("Tente novamente. Digite um número: "))
    finally:
        print("Obrigado!")
    print(val)
```

```python
askit()
```

```python
def askint():
    while True:
        try:
            val = int(input("Digite um número: "))
        except:
            print("Você não digitou um número!")
            continue
        else:
            print("Obrigado por digitar um número!")
            break
        finally:
            print("Fim da Execução!")
        print(val)
```

```python
askint()
```
